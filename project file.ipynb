```python
dm.drop(['Date'],axis =1, inplace = True)
```

```python
dm
```

```python
dm['HeightStart'].mean()
```

```python
x = dm["HeightStart"].mean()
dm["HeightStart"].fillna(x, inplace=True)
dm.round(0)
```

```python
dm['HeightEnd'].mean()
```

```python
x = dm["HeightEnd"].mean()
dm["HeightEnd"].fillna(x, inplace=True)
dm.round(0)
```

```python
dm.dropna(inplace = True)
```

```python
dm.isnull().sum()
```

```python
is_huge = np.logical_and(dm['HeightStart'] >250, dm['HeightEnd'] >300)
is_huge
```

```python
sns.swarmplot(x='Built', y='HeightStart', data=dm)
```
