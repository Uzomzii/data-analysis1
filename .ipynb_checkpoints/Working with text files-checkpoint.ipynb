### Importing Data with Numpy - np.loadtxt() vs np.genfromtxt()

```python
import numpy as np
```

```python
data = "Lending-Company-Numeric-Data.csv"
lend_co_data_num_1 = np.loadtxt(data, delimiter=',')

lend_co_data_num_1
```

```python
data = "Lending-Company-Numeric-Data.csv"
lend_co_data_num_2 = np.genfromtxt(data, delimiter=',')

lend_co_data_num_2
```

```python
np.array_equal(lend_co_data_num_2,lend_co_data_num_1)
```

```python
data2 = "Lending-Company-Numeric-Data-NAN.csv"
lend_co_data_num_1_nan = np.loadtxt(data2, delimiter=';', dtype=np.str)
lend_co_data_num_1_nan
```

```python
data2 = "Lending-Company-Numeric-Data-NAN.csv"
lend_co_data_num_2_nan = np.genfromtxt(data2, delimiter=';')
lend_co_data_num_2_nan
```

```python
filename2 = pd.read_csv('Lending-company.csv', usecols=['CustomerGender'])
filename2.head()
```

```python
type(filename2)
```

```python
df_single_col = pd.read_csv('Lending-company.csv', usecols=['CustomerGender'], squeeze=True)
df_single_col.head()
```

### Saving Files with NumPy - np.save()

```python
import numpy as np
```

```python
lending_co = np.genfromtxt("Lending-Company-Saving.csv",
                          delimiter = ',',
                          dtype = str)
print(lending_co)
```

```python
np.save('Saved-lending-file', lending_co)
```

### Saving Files with NumPy - np.savez()

```python
saved_data = np.load('Saved-lending-file.npy')
print(saved_data)
```

```python
np.savez('Lending-saved', saved_data)
```

### Saving Files with NumPy - np.savetxt()

```python
np.savetxt('Lending-text-file.txt',
          lending_co,
          fmt='%s',
          delimiter=',')
# We must specify the file extension (txt or csv)
# We must specify the format (strings in this case)
# We must set a delimiter (comma in this case)
```
